```python
import pandas as pd

PV_DATA_PATH = "../../data/rooftop_pv/"
DEMAND_DATA_PATH = "../../data/UNSW-PROJECT-DATA/"
```

```python
# Load the MEASUREMENT state-level data files
meas_df = pd.read_csv(PV_DATA_PATH+"rooftop_pv_measurement_wholestate.csv")

data_df = pd.read_csv(DEMAND_DATA_PATH+"combined_fd_Avg.csv")
```

```python
meas_df.head(10)
```

```python
# Drop the TAS1 data as that's not part of our analysis
meas_df = meas_df[meas_df['REGIONID'] != 'TAS1']

# Convert REGIONID to STATE to match the combined demand/forecast file
meas_df['STATE'] = meas_df['REGIONID'].map(lambda x: x[:-1])

# Now drop the un-needed I, ROOFTOP, ACTUAL, 2, QI, TYPE, I, LASTCHANGED and REGIONID columns
meas_df = meas_df.drop(['Unnamed: 0','ROOFTOP','ACTUAL','2','QI','TYPE','LASTCHANGED','I', 'REGIONID'], axis=1)
meas_df.head()
```

```python
meas_df.shape
```

```python
data_df.head(10)
```

```python
# Join the PV and combined demand/forecast dataframes by timestamp and state. 
# REGIONID has been modified by dropping the last character to give STATE in data_df so this needs to be takein into account
# There are only values for 2021 onwards in the PV data, so left-join on that to avoid 8 years of missing values.
data_df = data_df.drop('Unnamed: 0', axis=1)
result = pd.merge(meas_df, data_df, left_on=['STATE','INTERVAL_DATETIME'], right_on=['STATE','DATETIME'], how='left')
```

```python
# Write the combined datafile out as CSV
result.to_csv(PV_DATA_PATH+"pv_demand_forecast_combined.csv",index=False)
```
