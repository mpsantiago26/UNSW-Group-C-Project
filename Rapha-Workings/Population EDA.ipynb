```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import utils
```

```python
#Import population dataset
usecols = 'A,B,C,D,G,H,I,J,K,L'
skiprows = [1,2,3,4,5,6,7,8,9]
columns = ['Period', 'Births', 'Deaths', 'Natural Increase', 'Overseas Arrivals', 'Overseas Departures', 'Net Movement',
          'Migration Adjustment', 'Net Oversees Migration', 'Estimated Residential Population']
population = pd.read_excel("../../Datasets/population/population - National.xlsx", sheet_name="Data1", header=0, usecols=usecols, 
                           skiprows=skiprows)
population.columns = columns
```

```python
#Basic Statistics

population
```

```python
# Show Australia population trend from 1980
plt.figure(figsize = (15,8))
sns.lineplot(y='Estimated Residential Population', x='Period', data=population)
plt.xlabel('Reporting Period')
plt.ylabel('Population')
plt.title('Estimated Residentail Population (000)')
```

```python
# Show Australia population trend from 2012
plt.figure(figsize = (15,8))
sns.lineplot(y='Estimated Residential Population', x='Period', data=population[population.Period > '2012-01-01'])
plt.xlabel('Reporting Period')
plt.ylabel('Population')
plt.title('Estimated Residentail Population (000)')
```

```python
#Import population by state dataset
usecols = 'A,B,C,D,E,K,L,M,N,T,U,V,W'
skiprows = [1,2,3,4,5,6,7,8,9]
columns = ['Period', 'ERP(M) NSW', 'ERP(M) VIC', 'ERP(M) QLD', 'ERP(M) SA', 'ERP(F) NSW', 'ERP(F) VIC',
          'ERP(F) QLD', 'ERP(F) SA', 'ERP NSW', 'ERP VIC', 'ERP QLD', 'ERP SA']
population_states = pd.read_excel("../../Datasets/population/population - States.xlsx", sheet_name="Data1", header=0, usecols=usecols, 
                           skiprows=skiprows)
population_states.columns = columns
```

```python
#basic populationstatistics
population_states
```

```python
# Show Australia population trend from 2012
rows = population_states[population.Period > '2000-01-01'].index
cols = ['Period','ERP NSW', 'ERP VIC','ERP QLD','ERP SA']
fig = plt.figure(figsize = (15,8))
# sns.lineplot(data=population_states.loc[rows,cols])
sns.lineplot(x='Period',y='ERP VIC', data=population_states.loc[rows,cols])
sns.lineplot(x='Period',y='ERP NSW', data=population_states.loc[rows,cols])
sns.lineplot(x='Period',y='ERP SA', data=population_states.loc[rows,cols])
sns.lineplot(x='Period',y='ERP QLD', data=population_states.loc[rows,cols])
#plt.plot(x='Period',y='ERP VIC', data=population_states.loc[rows,cols])
plt.xlabel('Reporting Period')
plt.ylabel('Population')
plt.title('Estimated Residentail Population (000)')
fig.legend(labels=['VIC', 'NSW', 'SA', 'QLD'])
```

```python
#Load forecast data VIC
forecast_vic = pd.read_csv("../../Datasets/forecastdemand_vic.csv")
# Create extra fields from the data
forecast_vic['DATETIME'] = forecast_vic['DATETIME'].apply(lambda x: utils.calculate_month(x))
forecast_vic['YEAR'] = forecast_vic['DATETIME'].dt.year
forecast_vic['MONTH'] = forecast_vic['DATETIME'].dt.month_name().str[:3]
forecast_vic['SEASON'] = forecast_vic['MONTH'].apply(lambda x: utils.calculate_season(x))
```

```python
#Load demand data VIC
demand_vic = pd.read_csv("../../Datasets/totaldemand_vic.csv")
# Create extra fields from the data
demand_vic['DATETIME'] = demand_vic['DATETIME'].apply(lambda x: utils.calculate_month(x))
demand_vic['YEAR'] = demand_vic['DATETIME'].dt.year
demand_vic['MONTH'] = demand_vic['DATETIME'].dt.month_name().str[:3]
demand_vic['SEASON'] = demand_vic['MONTH'].apply(lambda x: utils.calculate_season(x))
```

```python
# Show forecast data for VIC
filter_date = '2019-01-01'
rows_f = forecast_vic[forecast_vic.DATETIME > filter_date].index
cols_f = ['DATETIME','FORECASTDEMAND']
rows_d = demand_vic[demand_vic.DATETIME > filter_date].index
cols_d = ['DATETIME','TOTALDEMAND']
fig = plt.figure(figsize = (15,8))
sns.lineplot(x='DATETIME',y='FORECASTDEMAND', data=forecast_vic.loc[rows_f,cols_f])
sns.lineplot(x='DATETIME',y='TOTALDEMAND', data=demand_vic.loc[rows_d,cols_d])
plt.xlabel('Reporting Period')
plt.ylabel('Demand Forecast')
plt.title('Demand Forecast VIC ')
fig.legend(labels=['FORECAST', 'DEMAND'])
```
